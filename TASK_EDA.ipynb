```python
def plot_stacked_bars(dataframe,title_,size_=(18,10),rot_=0,legend_="upper right"):
    """
    Plot STACKED BARS with annotations
    """
    ax=dataframe.plot(
    kind="bar",
    stacked=True,
    figsize=size_,
    rot=rot_,
    title=title_)
    
    #annotate bars
    annotate_stacked_bars(ax,textsize=18)
    #Rename the legend
    plt.legend(["Retention","Churn"],loc=legend_)
    plt.ylabel("Compant base (%)")
    plt.show()
    
def annotate_stacked_bars(ax,pad=0.99,colour="white",textsize=13):
    """
    Add values to annotated bars
    """
    for p in ax.patches:
        
        #Calculate the annotation
        value=str(round(p.get_height(),1))
        # if value is 0 do not annotate
        if value=="0":
            continue
        ax.annotate(
         value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )
```

# ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad)
#This line of code is part of a function called annotate_stacked_bars which is used to add value annotations to the stacked bars in a bar plot. Let's break down the code and explain each component:

p.get_x(): This method returns the x-coordinate of the bottom left corner of the bar rectangle.
p.get_width(): This method returns the width of the bar rectangle.
p.get_y(): This method returns the y-coordinate of the bottom left corner of the bar rectangle.
p.get_height(): This method returns the height of the bar rectangle.
pad: This variable is used to adjust the position of the annotation text.
-0.05: This constant value is subtracted from the x-coordinate to fine-tune the horizontal position of the annotation text.
*pad: This multiplication adjusts the position of the annotation text vertically.
Putting it all together, the line of code calculates the coordinates for placing the annotation text within each bar of the stacked bar plot. The x-coordinate is calculated as the midpoint of the bar (i.e., (p.get_x() + p.get_width()/2)), multiplied by pad and then subtracted by 0.05 to adjust the horizontal position. The y-coordinate is calculated as the midpoint of the bar (i.e., (p.get_y() + p.get_height()/2)), multiplied by pad to adjust the vertical position.

By using these calculated coordinates, the annotate_stacked_bars function adds the value annotations to the bars in the stacked bar plot, providing additional information about the height of each bar.

Overall, this line of code plays a crucial role in positioning the value annotations accurately within the stacked bars of the bar plot, enhancing the visual representation of the data.

#### for p in ax.patches:: This line initiates a loop that iterates over each patch (bar) in the ax object, which represents the bar chart.

value = str(round(p.get_height(),1)): This line calculates the value to be annotated for each bar. It retrieves the height of the current bar using p.get_height() and rounds it to one decimal place using the round() function. The value is then converted to a string using str().

if value == '0.0': continue: This line checks if the calculated value is equal to '0.0'. If it is, the code skips the annotation for that bar and continues to the next iteration of the loop. This is done to avoid annotating bars with a value of zero, as it may clutter the chart.

ax.annotate(value, ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad), color=colour, size=textsize): This line adds the annotation to the current bar. It uses the ax.annotate() function to place the annotation text (value) at a specific position on the chart. The position is calculated using the coordinates of the bar (p.get_x() and p.get_y()) and the width and height of the bar (p.get_width() and p.get_height()). The pad parameter adjusts the position of the annotation, and the color and size parameters set the color and size of the annotation text, respectively.

By using this code, the plot_stacked_bars function can generate a stacked bar chart with value annotations for each bar, providing additional information about the data being visualized.

It is important to note that this code assumes the existence of the ax object, which represents the axes of the bar chart. The ax object is created when the plot_stacked_bars function is called and is not shown in the provided code snippet.

This operation calculates the proportion of each value relative to the total sum of its row. The axis=0 parameter specifies that the division should be performed along the rows.

```python
def plot_distribution(dataframe, column, ax,bins_=50):
    """
    Plot distribution curver
    """
    
    #create a temporal dataframe with the data to be plotted
    temp=pd.DataFrame({"Retention":dataframe[dataframe["churn"]==0 ][column],"Churn": dataframe[dataframe["churn"]==1][column]})
    #plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

```python
fig, axs = plt.subplots(nrows=4, figsize=(18, 25))

plot_distribution(consumption, 'cons_12m', axs[0])
plot_distribution(consumption[consumption['has_gas'] == 't'], 'cons_gas_12m', axs[1])
plot_distribution(consumption, 'cons_last_month', axs[2])
plot_distribution(consumption, 'imp_cons', axs[3])
```

## HYPOTHESIS TESTING

####  Hypothesis investigation
Now that we have explored the data, it's time to investigate whether price sensitivity has some influence on churn. First we need to define exactly what is price sensitivity.

> Since we have the consumption data for each of the companies for the year of 2015, we will create new features to measure "price sensitivity" using the average of the year, the last 6 months and the last 3 months

```python
# Transform date columns to datetime type
client_df["date_activ"] = pd.to_datetime(client_df["date_activ"], format='%Y-%m-%d')
client_df["date_end"] = pd.to_datetime(client_df["date_end"], format='%Y-%m-%d')
client_df["date_modif_prod"] = pd.to_datetime(client_df["date_modif_prod"], format='%Y-%m-%d')
client_df["date_renewal"] = pd.to_datetime(client_df["date_renewal"], format='%Y-%m-%d')
price_df['price_date'] = pd.to_datetime(price_df['price_date'], format='%Y-%m-%d')

```

```python
price_df
```

```python
# Create mean average data
mean_year = price_df.groupby(['id']).mean().reset_index()
mean_6m = price_df[price_df['price_date'] > '2015-06-01'].groupby(['id']).mean().reset_index()
mean_3m = price_df[price_df['price_date'] > '2015-10-01'].groupby(['id']).mean().reset_index()
```

```python
mean_year
```

```python
# Comnbine into single dataframe
mean_year = mean_year.rename(
    index=str, 
    columns={
        "price_p1_var": "mean_year_price_p1_var",
        "price_p2_var": "mean_year_price_p2_var",
        "price_p3_var": "mean_year_price_p3_var",
        "price_p1_fix": "mean_year_price_p1_fix",
        "price_p2_fix": "mean_year_price_p2_fix",
        "price_p3_fix": "mean_year_price_p3_fix"
    }
)
```

```python
mean_year
```

```python
price_df.info()
```

```python
# Comnbine into single dataframe
mean_year = mean_year.rename(
    index=str, 
    columns={
        "price_off_peak_var": "mean_year_price_p1_var",
        "price_peak_var": "mean_year_price_p2_var",
        "price_mid_peak_var": "mean_year_price_p3_var",
        "price_off_peak_fix": "mean_year_price_p1_fix",
        "price_peak_fix": "mean_year_price_p2_fix",
        "price_mid_peak_fix": "mean_year_price_p3_fix"
    }
)
```

```python
mean_year["mean_year_price_p1"] = mean_year["mean_year_price_p1_var"] + mean_year["mean_year_price_p1_fix"]
mean_year["mean_year_price_p2"] = mean_year["mean_year_price_p2_var"] + mean_year["mean_year_price_p2_fix"]
mean_year["mean_year_price_p3"] = mean_year["mean_year_price_p3_var"] + mean_year["mean_year_price_p3_fix"]
```

```python
mean_6m = mean_6m.rename(
    index=str, 
    columns={
        "price_off_peak_var": "mean_6m_price_p1_var",
        "price_peak_var": "mean_6m_price_p2_var",
        "price_mid_peak_var": "mean_6m_price_p3_var",
       "price_off_peak_fix": "mean_6m_price_p1_fix",
        "price_peak_fix": "mean_6m_price_p2_fix",
        "price_mid_peak_fix": "mean_6m_price_p3_fix"
    }
)
mean_6m["mean_6m_price_p1"] = mean_6m["mean_6m_price_p1_var"] + mean_6m["mean_6m_price_p1_fix"]
mean_6m["mean_6m_price_p2"] = mean_6m["mean_6m_price_p2_var"] + mean_6m["mean_6m_price_p2_fix"]
mean_6m["mean_6m_price_p3"] = mean_6m["mean_6m_price_p3_var"] + mean_6m["mean_6m_price_p3_fix"]

mean_3m = mean_3m.rename(
    index=str, 
    columns={
        "price_off_peak_var": "mean_3m_price_p1_var",
         "price_peak_var": "mean_3m_price_p2_var",
        "price_mid_peak_var": "mean_3m_price_p3_var",
        "price_off_peak_fix": "mean_3m_price_p1_fix",
        "price_peak_fix": "mean_3m_price_p2_fix",
        "price_mid_peak_fix": "mean_3m_price_p3_fix"
    }
)
mean_3m["mean_3m_price_p1"] = mean_3m["mean_3m_price_p1_var"] + mean_3m["mean_3m_price_p1_fix"]
mean_3m["mean_3m_price_p2"] = mean_3m["mean_3m_price_p2_var"] + mean_3m["mean_3m_price_p2_fix"]
mean_3m["mean_3m_price_p3"] = mean_3m["mean_3m_price_p3_var"] + mean_3m["mean_3m_price_p3_fix"]

# Merge into 1 dataframe
price_features = pd.merge(mean_year, mean_6m, on='id')
price_features = pd.merge(price_features, mean_3m, on='id')
```

```python
price_features.head()
```

Now lets merge in the churn data and see whether price sensitivity has any correlation with churn

```python
price_analysis = pd.merge(price_features, client_df[['id', 'churn']], on='id')
price_analysis.head()
```

```python
corr = price_analysis.corr()
# Plot correlation
plt.figure(figsize=(20,18))
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, annot = True, annot_kws={'size':10})
# Axis ticks size
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

From the correlation plot, it shows a higher magnitude of correlation between other price sensitivity variables, however overall the correlation with churn is very low. This indicates that there is a weak linear relationship between price sensitity and churn. This suggests that for price sensivity to be a major driver for predicting churn, we may need to engineer the feature differently.
